Question 4: Describe the architecture and functioning of a Recurrent Natural Network (RNN).

Ans: A Recurrent Neural Network (RNN) is a class of artificial neural networks designed to process sequential data. Unlike traditional feed-forward neural networks, which assume all inputs are independent of each other, RNNs possess a "memory" that allows them to retain information from previous steps in a sequence.This makes them the foundational architecture for tasks like language translation, speech recognition, and time-series forecasting.1. The Core Architecture: The Feedback LoopThe defining feature of an RNN is the recurrent connection. In a standard network, data flows in one direction: from input to output. In an RNN, the output of a hidden layer at a specific time step is cycled back as an input for the next time step.Input ($x_t$): The data point at the current time (e.g., a single word in a sentence).Hidden State ($h_t$): The "memory" of the network. It is calculated based on the current input and the previous hidden state.Output ($y_t$): The prediction made at the current step.2. How It Functions: "Unrolling" Through TimeTo understand how an RNN processes a sequence, it is helpful to "unroll" the loop. Imagine you are processing the sentence "The cat sat."Step 1 ($t=1$): The network takes the word "The" ($x_1$). It generates a hidden state ($h_1$) representing this word.Step 2 ($t=2$): The network takes the word "cat" ($x_2$) and the previous hidden state ($h_1$). The new hidden state ($h_2$) now contains information about both "The" and "cat."Step 3 ($t=3$): The word "sat" ($x_3$) is combined with ($h_2$). The final output ($y_3$) is influenced by the entire context of the sentence.The mathematical representation of the hidden state is typically:ht​=σ(Wh​ht−1​+Wx​xt​+b)Where $W$ represents weight matrices, $b$ is the bias, and $\sigma$ is an activation function (usually tanh or ReLU).3. Training: Backpropagation Through Time (BPTT)RNNs are trained using a variation of backpropagation called Backpropagation Through Time (BPTT).The error (loss) is calculated at each time step.Because the parameters (weights) are shared across all time steps, the gradient is calculated by summing the errors across the entire sequence length.4. The Main Limitation: Vanishing GradientsWhile RNNs are conceptually powerful, they struggle with long-term dependencies.As a sequence gets longer, the gradients used to update the weights during BPTT can become extremely small (the Vanishing Gradient problem) or extremely large (Exploding Gradients).This means the network "forgets" information from the beginning of a long paragraph by the time it reaches the end.Summary TableFeatureDescriptionInput TypeSequential (Text, Audio, Stock Prices)Key MechanismHidden State (Internal Memory)Weight SharingSame weights are used for every step in the sequenceMajor WeaknessDifficulty remembering long-term contextSuccessorsLSTMs (Long Short-Term Memory) and GRUs

Question 5: What is the key difference between LSTM and GRU networks in NLP applications?

Ans:While both LSTMs (Long Short-Term Memory) and GRUs (Gated Recurrent Units) were designed to solve the "vanishing gradient" problem of standard RNNs, they differ primarily in their structural complexity and how they manage their internal memory.In NLP, these differences often translate to a trade-off between precision and computational efficiency.1. The Architectural DifferenceThe fundamental distinction lies in the number of "gates" used to control the flow of information.LSTM: The Three-Gate SystemAn LSTM uses a complex architecture with three gates and a separate cell state ($c_t$).Forget Gate: Decides what information from the previous state should be discarded.Input Gate: Decides which new information from the current input should be stored in the cell state.Output Gate: Decides what part of the cell state should be output as the hidden state.Cell State ($c_t$): Acts like a "conveyor belt," carrying long-term information through the entire sequence with minimal interference.GRU: The Two-Gate SystemA GRU is a simplified version of the LSTM. It merges the cell state and the hidden state and reduces the architecture to two gates.Update Gate: Determines how much of the previous memory to keep and how much new information to add (combining the roles of the LSTM's forget and input gates).Reset Gate: Determines how much of the past information to forget.Simplicity: Because it has fewer parameters, a GRU is computationally "cheaper" and faster to train than an LSTM.2. Performance in NLP ApplicationsBecause human language is nuanced and often contains long-distance dependencies, the choice between them matters:FeatureLSTMGRUComplexityHigher (More parameters)Lower (Fewer parameters)MemorySeparate Cell and Hidden statesCombined Hidden stateTraining SpeedSlowerFasterData RequirementPerforms better with large datasetsCan be more efficient with smaller datasetsNLP StrengthExcellent for very long sentences/booksGreat for real-time speech/shorter text3. Which one is "Better"?There is no universal winner, but here are the general trends in the industry:Use LSTM if you are dealing with very long-range dependencies (e.g., long-form document summarization) and have ample computing power. Its separate cell state makes it slightly more robust at "remembering" things over very long sequences.Use GRU if you need to iterate quickly or are working with limited hardware. In many modern NLP benchmarks, GRUs achieve performance nearly identical to LSTMs but with significantly faster training times and less memory usage.Fun Fact: Most modern "state-of-the-art" NLP has moved toward Transformers (like BERT or GPT), but LSTMs and GRUs are still widely used in "Edge AI" (like on your phone) because they are much lighter than massive Transformer models.

the following paragraph of text:
“I had a great experience using the new mobile banking app. The interface is intuitive,
and customer support was quick to resolve my issue. However, the app did crash once

Your program should print out the polarity and subjectivity scores.

Ans: To perform sentiment analysis with TextBlob, you first need to ensure the library is installed (pip install textblob). The library makes it incredibly simple to extract emotional nuances from a string of text by simply calling the .sentiment attribute.

Here is the Python code to analyze your provided paragraph:

Python
from textblob import TextBlob

# The input paragraph
text = """I had a great experience using the new mobile banking app.
The interface is intuitive, and customer support was quick to resolve my issue.
However, the app did crash once during a transaction, which was frustrating."""

# Create a TextBlob object
blob = TextBlob(text)

# Extract sentiment scores
# Polarity ranges from -1 (negative) to 1 (positive)
# Subjectivity ranges from 0 (objective) to 1 (subjective)
polarity = blob.sentiment.polarity
subjectivity = blob.sentiment.subjectivity

print(f"Text: {text}\n")
print("-" * 30)
print(f"Polarity Score: {polarity:.2f}")
print(f"Subjectivity Score: {subjectivity:.2f}")
print("-" * 30)

# Quick Interpretation
if polarity > 0:
    print("Overall Sentiment: Positive")
elif polarity < 0:
    print("Overall Sentiment: Negative")
else:
    print("Overall Sentiment: Neutral")

Python
import nltk
from nltk.tokenize import word_tokenize
from nltk.probability import FreqDist
import string

# 2. Cleaning (Optional but recommended: removing punctuation)

print("-" * 30)
print("Top 5 Most Frequent Words:")
for word, frequency in fdist.most_most_common(5):

Question 8: Implement a basic LSTM model in Keras for a text classification task using

(1) or negative (0).
# Dataset
texts = [
“I love this project”, #Positive
“This is an amazing experience”, #Positive
“I hate waiting in line”, #Negative
“This is the worst service”, #Negative
“Absolutely fantastic!” #Positive
]
labels = [1, 1, 0, 0, 1]
Preprocess the text, tokenize it, pad sequences, and build an LSTM model to train on
this data. You may use Keras with TensorFlow backend.
(Include your Python code and output in the code box below.)

Ans: To implement an LSTM in Keras, we follow a specific pipeline: tokenization (turning words into numbers), padding (making all sequences the same length), and then defining the model architecture with an Embedding layer followed by the LSTM.

Here is the complete implementation using the dummy dataset provided:

Python
import numpy as np
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense

# 1. Dataset
texts = [
    "I love this project",
    "This is an amazing experience",
    "I hate waiting in line",
    "This is the worst service",
    "Absolutely fantastic!"
]
labels = np.array([1, 1, 0, 0, 1])

# 2. Preprocessing & Tokenization
# We set a small vocabulary size for this dummy data
max_words = 1000  
tokenizer = Tokenizer(num_words=max_words, oov_token="<OOV>")
tokenizer.fit_on_texts(texts)
sequences = tokenizer.texts_to_sequences(texts)

# 3. Padding sequences (ensuring all inputs have the same length)
max_len = 10
padded_sequences = pad_sequences(sequences, maxlen=max_len, padding='post')

# 4. Building the LSTM Model
model = Sequential([
    # Embedding layer: converts word indices into dense vectors
    Embedding(input_dim=max_words, output_dim=16, input_length=max_len),
    # LSTM layer with 32 units
    LSTM(32),
    # Output layer: Sigmoid activation for binary classification (0 or 1)
    Dense(1, activation='sigmoid')
])

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# 5. Training (Small number of epochs for dummy data)
print("Starting training...")
history = model.fit(padded_sequences, labels, epochs=10, verbose=0)

# 6. Evaluation on the same data
loss, accuracy = model.evaluate(padded_sequences, labels, verbose=0)

print("-" * 30)
print(f"Final Training Accuracy: {accuracy * 100:.2f}%")
print("-" * 30)

# 7. Testing a new sentence
test_text = ["I love this service"]
test_seq = tokenizer.texts_to_sequences(test_text)
test_padded = pad_sequences(test_seq, maxlen=max_len)
prediction = model.predict(test_padded, verbose=0)
sentiment = "Positive" if prediction[0][0] > 0.5 else "Negative"

print(f"Test Sentence: '{test_text[0]}'")
print(f"Predicted Sentiment: {sentiment} (Score: {prediction[0][0]:.4f})")

Key Components Explained
Tokenizer: This maps each unique word to a unique integer. For instance, the word "love" might be represented by the number 3. This is necessary because neural networks cannot process raw text.

pad_sequences: LSTM layers require a fixed input shape. Since sentences vary in length, we add zeros (padding) to the end of shorter sentences so they all reach a uniform length of 10.

Embedding Layer: This is a critical first step in NLP. It learns a dense vector representation for each word, ensuring that words with similar meanings (like "amazing" and "fantastic") are mathematically closer to each other.

LSTM Layer: This is the "brain" that captures the sequential relationship between words, allowing the model to understand context rather than just individual word counts.

Sigmoid Activation: Used in the final Dense layer to squash the output between 0 and 1. A value closer to 1 indicates a positive sentiment, while closer to 0 indicates negative.

import spacy
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense, GRU
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.preprocessing.text import Tokenizer

# Load spaCy model
nlp = spacy.load("en_core_web_sm")

# Sample dataset (user input → label)
texts = [
    "I feel very sad and lonely",
    "I am anxious about my future",
    "I am happy today",
    "I feel depressed and tired"
]

labels = [0, 1, 2, 0]  # 0=sad, 1=anxious, 2=happy

# Preprocessing using spaCy
def preprocess(text):
    doc = nlp(text.lower())
    tokens = [token.lemma_ for token in doc if not token.is_stop and token.is_alpha]
    return " ".join(tokens)

processed_texts = [preprocess(t) for t in texts]

# Tokenization
tokenizer = Tokenizer()
tokenizer.fit_on_texts(processed_texts)
sequences = tokenizer.texts_to_sequences(processed_texts)

# Padding
X = pad_sequences(sequences, maxlen=10)

y = np.array(labels)

# Build LSTM model
model = Sequential()
model.add(Embedding(input_dim=1000, output_dim=64, input_length=10))
model.add(LSTM(64))  # or replace with GRU(64)
model.add(Dense(32, activation='relu'))
model.add(Dense(3, activation='softmax'))

model.compile(loss='sparse_categorical_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])

# Train model
model.fit(X, y, epochs=10, verbose=1)

# Test input
test_text = "I feel very low and sad"
processed_test = preprocess(test_text)
seq = tokenizer.texts_to_sequences([processed_test])
padded = pad_sequences(seq, maxlen=10)

prediction = model.predict(padded)

print("Processed Input:", processed_test)
print("Predicted Class:", np.argmax(prediction))
